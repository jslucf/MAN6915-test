```python
def encrypt(rotation):
    #encrypt a message based on a user-inputted value to rotate on the character map
    
    #prompt user for message
    msg = input("What is the message? ")
    
    #append each letter as cipher number in message into new list, msg_letters
    msg_letters = []
    for i in msg:
        msg_letters.append(ord(i))
    
    msg_rotate = []
    #now convert number back into character added to rotation number
    for i in msg_letters:
        msg_rotate.append(chr(i + rotation))
    
    #turn list of characters into one string and print
    msg_encrypt = "".join(msg_rotate)
    return msg_encrypt
    
    
def decrypt(msg_encrypt, rotation):
    #decrypt an encrypted string back
    
    #convert encrypted letters back into numbers with the rotation reversed
    msg_decrypt1 = []
    for i in msg_encrypt:
        msg_decrypt1.append(ord(i) - rotation)
     
    #Now convert numbers back to characters in one list
    msg_decrypt2 = []
    for i in msg_decrypt1:
        msg_decrypt2.append(chr(i))
    
    #join all the decrypted characters back into 1
    msg_decrypt3 = "".join(msg_decrypt2)
    return msg_decrypt3

def encr_decr(rotation):
    #use the msg_encrypt and msg_decrypt functions to scramble and unscramble a user message
    
    #take user message and encrypt
    encr = encrypt(rotation)
    
    #decrypt the encrypted msg
    decr = decrypt(encr,rotation)
    
    #print to the console
    print("Your encrypted message is: {}".format(encr))
    print("Your decrypted message is: {}".format(decr))
    
encr_decr(13)
encr_decr(47)
encr_decr(-13)
```
